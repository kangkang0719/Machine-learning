# 模块导入和数据导入

```python
#基础函数库
import numpy as np 
import pandas as pd

#绘图函数库
import matplotlib.pyplot as plt
import seaborn as sns

#模型相关函数库
from sklearn.model_selection import train_test_split
from lightgbm.sklearn import LGBMClassifier
from sklearn import metrics
from lightgbm import plot_importance
from sklearn.model_selection import GridSearchCV
```

```python
#数据导入
data=pd.read_csv(r"C:\Users\99036\Desktop\杂七杂八文档\机械学习作业\综合实践\英雄联盟数据集.csv")
data
```

```python
#查看整体数据类型
data.info()
```

# 数据预处理

```python
#获取标签
y=data.blueWins
y.value_counts()
```

```python
#删除标签和游戏场次id查看特征的数据特征
drop_cols = ['gameId','blueWins']
temp=data.drop(drop_cols, axis=1)
temp.describe()
```

```python
#删除不必要特征
drop_cols1=['blueDragons','blueHeralds','redFirstBlood','redKills','redDeaths','redDragons','redHeralds','redGoldDiff','redExperienceDiff']
temp1=temp.drop(drop_cols1,axis=1)
temp1.describe()
```

```python
#绘制小提琴图对数据进行分析
v_data = temp1
v_data_std = (v_data - v_data.mean()) / v_data.std()
v_data = pd.concat([y, v_data_std.iloc[:, 0:9]], axis=1)
v_data = pd.melt(v_data, id_vars='blueWins', var_name='Features', value_name='Values')

fig, ax = plt.subplots(1,2,figsize=(15,5))

# 绘制小提琴图1
sns.violinplot(x='Features', y='Values', hue='blueWins', data=v_data, split=True,
               inner='quart', ax=ax[0], palette='Blues')
fig.autofmt_xdate(rotation=45)

v_data1 = pd.concat([y, v_data_std.iloc[:, 9:17]], axis=1)
v_data1 = pd.melt(v_data1, id_vars='blueWins', var_name='Features', value_name='Values')

# 绘制小提琴图2
sns.violinplot(x='Features', y='Values', hue='blueWins', data=v_data1, split=True,
               inner='quart', ax=ax[1], palette='Blues')
fig.autofmt_xdate(rotation=45)

plt.show()
```

```python
#查看各特征之间的相关性
plt.figure(figsize=(18,14))
sns.heatmap(round(temp1.corr(),2), cmap='Blues', annot=True)
plt.show()
```

```python
#删除冗余特征
drop_cols2=['redAvgLevel','blueAvgLevel','blueCSPerMin','redCSPerMin','blueGoldPerMin','redGoldPerMin']
temp2=temp1.drop(drop_cols2,axis=1)
temp2.describe()
```

```python
#去除和眼位相关的特征
drop_cols3= ['blueWardsPlaced','blueWardsDestroyed','redWardsPlaced','redWardsDestroyed']
temp3=temp2.drop(drop_cols3,axis=1)
temp3.describe()
```

```python
#将一些特征合并
temp3['killsDiff'] = temp3['blueKills'] - temp3['blueDeaths']
temp3['assistsDiff'] = temp3['blueAssists'] - temp3['redAssists']
#查看合并前后的击杀、助攻、和其差的分布图
temp3[['blueKills','blueDeaths','blueAssists','killsDiff','assistsDiff','redAssists']].hist(figsize=(12,10), bins=20)
plt.show()
```

```python
#去除击杀和助攻合并前的特征
drop_cols4= ['blueKills','blueDeaths','blueAssists','redAssists']
temp4=temp3.drop(drop_cols4,axis=1)
temp4.describe()
```

```python
#探究塔对胜率的影响
temp4['towerDiff'] = temp4['blueTowersDestroyed'] - temp4['redTowersDestroyed']

g_data = pd.concat([y, temp4], axis=1)

towerGroup = g_data.groupby(['towerDiff'])['blueWins']
print(towerGroup.count())
print(towerGroup.mean())

fig, ax = plt.subplots(1,2,figsize=(15,5))

towerGroup.mean().plot(kind='line', ax=ax[0])
ax[0].set_title('Proportion of Blue Wins')
ax[0].set_ylabel('Proportion')

towerGroup.count().plot(kind='line', ax=ax[1])
ax[1].set_title('Count of Towers Destroyed')
ax[1].set_ylabel('Count')
```

# 模型建立

```python
#划分测试集和训练集
data_target_part = y
data_features_part = temp4
#训练和测试集比例为4：1
x_train, x_test, y_train, y_test = train_test_split(data_features_part, data_target_part, test_size = 0.2, random_state = 2023)
```

```python
#LightGBM模型
clf = LGBMClassifier()
#在训练集上训练LightGBM模型
clf.fit(x_train, y_train)
#在训练集和测试集上分布利用训练好的模型进行预测
train_predict = clf.predict(x_train)
test_predict = clf.predict(x_test)
print('训练集上的预测准确度:',metrics.accuracy_score(y_train,train_predict))
print('测试集上的预测准确度:',metrics.accuracy_score(y_test,test_predict))
```

```python
#查看混淆矩阵
confusion_matrix_result = metrics.confusion_matrix(test_predict,y_test)
print('测试集上的预测结果的混淆矩阵:\n',confusion_matrix_result)
```

```python
# 利用热力图对于结果进行可视化
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix_result, annot=True, cmap='Blues')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

```python
#查看各个特征的重要度
sns.barplot(y=data_features_part.columns, x=clf.feature_importances_)
```

```python
#使用plot_importance来查看特征的重要性
#使用ligthgbm中两种常用计算方法计算特征的重要度
#gain：当利用特征做划分的时候的评价基尼指数
#split：是以特征用到的次数来评价
def estimate(model,data):
    ax1=plot_importance(model,importance_type="gain")
    ax1.set_title('gain')
    ax2=plot_importance(model, importance_type="split")
    ax2.set_title('split')
    plt.show()
def classes(data,label,test):
    model=LGBMClassifier()
    model.fit(data,label)
    ans=model.predict(test)
    estimate(model, data)
    return ans
 
ans=classes(x_train,y_train,x_test)
```

```python
#我们选择重要性最高的前11个特征进行重新训练
data_choose=temp4[['redTotalExperience','blueGoldDiff','blueTotalGold','redTotalMinionsKilled','blueTotalExperience','redTotalGold','blueTotalMinionsKilled','blueExperienceDiff','redTotalJungleMinionsKilled','blueTotalJungleMinionsKilled','assistsDiff']]
#划分测试集和训练集
data_target_part1 = y
data_features_part1 = data_choose
#训练和测试集比例为4：1
x_train1, x_test1, y_train1, y_test1 = train_test_split(data_features_part1, data_target_part1, test_size = 0.2, random_state = 2023)
```

```python
#查看混淆矩阵
confusion_matrix_result1 = metrics.confusion_matrix(test_predict1,y_test1)
print('测试集上的预测结果的混淆矩阵:\n',confusion_matrix_result1)
```

```python
## 定义参数取值范围
learning_rate = [0.1, 0.3, 0.6]
feature_fraction = [0.5, 0.8, 1]
num_leaves = [16, 32, 64]
max_depth = [-1,3,5,8]

parameters = { 'learning_rate': learning_rate,
              'feature_fraction':feature_fraction,
              'num_leaves': num_leaves,
              'max_depth': max_depth}
model_adj = LGBMClassifier(n_estimators = 50)
## 进行网格搜索
clf_GCV = GridSearchCV(model_adj, parameters, cv=3, scoring='accuracy',verbose=3, n_jobs=-1)
clf_GCV= clf_GCV.fit(x_train, y_train)
```

```python
## 网格搜索后的最好参数为
clf_GCV.best_params_
```
